### Ques9 . average weekly response time for each agent 

```python
performance
def convert(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
    return "%d:%02d:%02d" % (hour, minutes, seconds)

# converting time resolution in seconds
performance['Time_responce'] = pd.to_timedelta(performance['Average Response Time']).astype('timedelta64[s]').astype(int)
df9 = pd.DataFrame(performance[performance['Average Rating'] >  0] .groupby(['Agent Name', 'Week'])['Time_responce'].mean())

# Converting sec back into time 
df9['Time_response'] = df9['Time_responce'].apply(convert)
df9
```


### Ques10 . average weekely resolution time for each agents 

```python

# converting "Average resolution time" into time format
performance['Time_resolution'] = pd.to_timedelta(performance['Average Resolution Time']).astype('timedelta64[s]').astype(int)

df10 = pd.DataFrame(performance[performance['Average Rating'] > 0].groupby(['Agent Name', 'Week'])['Time_resolution'].mean())
df10['Time_resolution'] = df10['Time_resolution'].apply(convert)
df10
```
